# Q12. Feature Engineering

Create a new column 'TotalScore' by adding all score columns

Create a 'Grade' column based on CGPA: A (3.5-4.0), B (3.0-3.5), C (2.5-3.0), D (below 2.5)

Create a 'PassStatus' column: 'Pass' if FinalScore >= 50, else 'Fail'

Calculate the percentage of students who passed

```python
# Create a new column 'TotalScore' by adding all score columns
df["TotalScore"]=df['AssignmentScore']+df['MidtermScore']+df['FinalScore']+df['ProjectScore']
df
```

```python
# Create a 'Grade' column based on CGPA: A (3.5-4.0), B (3.0-3.5), C (2.5-3.0), D (below 2.5)
conditions = [df['CGPA']>= 3.5,
             df['CGPA']>=3.0,
             df['CGPA']>=2.5,
             df['CGPA']< 2.5  ]
choices=['A','B','C','D']
df["Grade"]=np.select(conditions,choices,default='D')
df['Grade']=np.where(df['CGPA']>=3.5,'A',
            np.where(df['CGPA']>=3.0,'B',
            np.where(df['CGPA']>=2.5,'C','D')))
(df)
```

```python
# Create a 'PassStatus' column: 'Pass' if FinalScore >= 50, else 'Fail'
conditions= [df['FinalScore']>=50, df['FinalScore']<50]
choices= ['Pass','Fail']
df["Pass_Status"]=np.select(conditions,choices,default='Fail')
df
```

```python
# Calculate the percentage of students who passed
df['Passed']=np.where(df['FinalScore']>=50,'Pass','Fail')
df
```

```python
# Calculate the percentage of students who passed
df1=df[df['Pass_Status']=='Pass']
df2=len(df1)
df3=(df2/len(df))*100
print("percentage of students who passed",df3)
```

```python
df.drop(['Passed'],axis=1,inplace=True)
```

**Q11. Statistical Analysis with NumPy**

Calculate the total score for each student (sum of AssignmentScore, MidtermScore, FinalScore, ProjectScore)

Use NumPy to find students whose total score is above the 80th percentile

Calculate the standard deviation of AttendancePercent using NumPy


```python
# Calculate the total score for each student (sum of AssignmentScore, MidtermScore, FinalScore, ProjectScore)
df['TotalScore'] = df[['AssignmentScore', 'MidtermScore', 'FinalScore', 'ProjectScore']].sum(axis=1)
df[['StudentID','TotalScore']]
```

```python
# Use NumPy to find students whose total score is above the 80th percentile
df1=np.array(df['TotalScore'])
p80=np.percentile(df1,80)
mask = df1 > p80
p80_a=df[mask]
print(p80)
print(p80_a)
```

```python
import numpy as np

# Convert TotalScore column to NumPy array
total = df['TotalScore'].to_numpy()

# Calculate the 80th percentile
p80 = np.percentile(total, 80)

# Get boolean mask of students above p80
mask = total > p80

# Filter the DataFrame
above_80th = df[mask]

print("80th Percentile:", p80)
print(above_80th)
```
