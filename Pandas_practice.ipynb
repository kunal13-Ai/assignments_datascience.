```python
import pandas as pd
import numpy as np

# Create a 7-dimensional India farming dataset for practice
np.random.seed(42)

# Dimension 1: Farmer Names (Indian names)
first_names = ['Ramesh', 'Suresh', 'Rajesh', 'Mahesh', 'Ganesh', 'Dinesh', 'Prakash',
               'Amit', 'Vijay', 'Ajay', 'Ravi', 'Krishna', 'Mohan', 'Gopal', 'Anil',
               'Lakshmi', 'Savita', 'Kamala', 'Radha', 'Sita', 'Gita', 'Priya', 'Meera']
last_names = ['Kumar', 'Singh', 'Patel', 'Sharma', 'Reddy', 'Rao', 'Nair', 'Desai',
              'Yadav', 'Joshi', 'Gupta', 'Verma', 'Chauhan', 'Pillai', 'Naik']
farmer_names = [f'{np.random.choice(first_names)} {np.random.choice(last_names)}'
                for _ in range(300)]

# Dimension 2: States/Regions in India
states = ['Punjab', 'Haryana', 'Uttar Pradesh', 'Madhya Pradesh', 'Maharashtra',
          'Karnataka', 'Tamil Nadu', 'Andhra Pradesh', 'West Bengal', 'Gujarat',
          'Rajasthan', 'Bihar', 'Odisha', 'Kerala', 'Telangana']

# Dimension 3: Crop Types
crops = [ 'Wheat', 'Cotton', 'Sugarcane', 'Maize', 'Groundnut',
         'Soybean', 'Millets',  'Mustard','Coconut', 'Potato',
          'Onion', 'Tomato']

# Dimension 4: Season
seasons = ['Kharif', 'Rabi', 'Zaid']  # Indian agricultural seasons

# Dimension 5: Year and Month (Planting Date)
years = [2021, 2022, 2023, 2024]
dates = pd.date_range('2021-01-01', '2024-12-31', freq='D')

# Generate 1500 sample farming records
data = {
    'farmer_name': np.random.choice(farmer_names, 1500),
    'state': np.random.choice(states, 1500),
    'crop': np.random.choice(crops, 1500),
    'season': np.random.choice(seasons, 1500),
    'planting_date': np.random.choice(dates, 1500),
    'area_hectares': np.random.uniform(0.5, 50, 1500).round(2),
    'duration_days': np.random.randint(60, 180, 1500)
}
df=pd.DataFrame(data)
print(df)
```
