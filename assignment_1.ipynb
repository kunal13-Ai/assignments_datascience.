12. Python program to find the smallest and largest number in a list (Without min-max function)


```python
# using .sort method
a=[21,26,15,27,35,49,7]
a.sort()
largest = a[-1]
smallest= a[0]
print('largest number:',largest)
print('smallest number:',smallest)
```

```python
# using for loop
a=[21,26,15,27,35,49,7]
largest=a[-2]
smallest=a[0]
for num in a:
  if num > largest:
    largest=num
  if num < smallest:
    smallest=num
print('largest number:',largest)
print('smallest number:',smallest)
```

```python
a=[20,51,34,62,19,39] #if there is no duplicate in given list.
a.sort()
c=a[3]
print("third largest number:",c)
```

```python
a=[20,51,34,62,19,39,51,34]
first=a[0] #to compare we have taken smallest no. possible
second=a[0]
third=a[0]
for num in a:
  if num > first: # shifting the no. when we find bigger no. than previous one
    third=second  # second become third
    second=first  # first become second
    first=num     # num become first
  elif num > second and num !=first : # this mean num is not equal to first one
    second=first
    second=num
  elif num > third and num !=second:
    third=num
print("third largest number:",third)
```
